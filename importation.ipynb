```python
# Interface de recherche interactive
def rechercher_contribuable_interactif():
    """Interface interactive pour rechercher un contribuable"""
    print("🔍 RECHERCHE DE CONTRIBUABLE")
    print("=" * 40)
    
    # Afficher quelques exemples d'IDs
    print("Exemples d'ID de contribuables:")
    sample_ids = df_srmt['ID_CONTRIBUABLE'].unique()[:10]
    for i, contrib_id in enumerate(sample_ids, 1):
        print(f"  {i}. {contrib_id}")
    
    contrib_id = input("\n👉 Entrez l'ID du contribuable à rechercher: ").strip()
    
    if contrib_id:
        info = agent.get_contribuable_info(contrib_id)
        
        if "error" in info:
            print(f"❌ {info['error']}")
        else:
            print(f"\n✅ CONTRIBUABLE TROUVÉ: {contrib_id}")
            print("-" * 50)
            print(f"• Pays: {info['pays']}")
            print(f"• Régime: {info['type_regime']}")
            print(f"• Direction: {info['direction']}")
            print(f"• Bureau: {info['bureau']}")
            print(f"• Déclarations: {info['nombre_declarations']}")
            print(f"• Montant déclaré: {info['montant_total_declare']:,.2f} FCFA")
            print(f"• Montant recouvré: {info['montant_total_recouvre']:,.2f} FCFA")
            print(f"• Taux recouvrement: {info['taux_recouvrement']:.2f}%")
            print(f"• Période: {info['periode_activite']['debut']} → {info['periode_activite']['fin']}")
            
            # Analyse IA automatique
            print(f"\n🤖 ANALYSE IA AUTOMATIQUE...")
            ai_analysis = agent.analyze_with_ai(
                f"Analyse le profil fiscal du contribuable {contrib_id} et donne tes recommandations.", 
                info
            )
            print("💡 Recommandations de l'IA:")
            print(ai_analysis)

# Lancer la recherche interactive
rechercher_contribuable_interactif()
```

```python
# 🎯 TOP 10 DES CONTRIBUABLES PAR MONTANT RECOUVRÉ - Version Auto-Corrective
print("🏆 ANALYSE TOP 10 CONTRIBUABLES - AUTO-CORRECTION INTÉGRÉE")
print("=" * 60)

def analyser_top_contribuables_auto_corrective():
    """Fonction avec auto-correction pour analyser les top contribuables"""
    try:
        # Étape 1: Vérifier les colonnes disponibles
        print("📋 Vérification des colonnes disponibles...")
        colonnes_montant = [col for col in df_srmt.columns if 'MONTANT' in col.upper()]
        print(f"Colonnes de montant détectées: {colonnes_montant}")
        
        # Étape 2: Analyser les types de données
        for col in colonnes_montant:
            dtype = df_srmt[col].dtype
            non_null = df_srmt[col].count()
            print(f"• {col}: {dtype} ({non_null:,} valeurs non-nulles)")
        
        # Étape 3: Tentative avec MONTANT_RECOUVRE
        if 'MONTANT_RECOUVRE' in df_srmt.columns:
            print("\n🔧 Tentative 1: Utilisation directe de MONTANT_RECOUVRE...")
            
            # Vérifier le type de données
            if df_srmt['MONTANT_RECOUVRE'].dtype == 'object':
                print("⚠️  Colonne MONTANT_RECOUVRE est de type 'object' - Conversion nécessaire")
                
                # Auto-correction: Conversion en numérique
                print("🛠️  AUTO-CORRECTION: Conversion en numérique...")
                df_work = df_srmt.copy()
                df_work['MONTANT_RECOUVRE'] = pd.to_numeric(df_work['MONTANT_RECOUVRE'], errors='coerce')
                df_work['MONTANT_RECOUVRE'] = df_work['MONTANT_RECOUVRE'].fillna(0)
                
                print(f"✅ Conversion réussie - {df_work['MONTANT_RECOUVRE'].count():,} valeurs numériques")
                
            else:
                df_work = df_srmt.copy()
                print("✅ Colonne déjà numérique")
            
            # Calcul du top 10 par contribuable
            print("\n📊 Calcul du top 10 par ID_CONTRIBUABLE...")
            top_contribuables = (df_work.groupby('ID_CONTRIBUABLE')['MONTANT_RECOUVRE']
                               .sum()
                               .sort_values(ascending=False)
                               .head(10))
            
            print("\n🏆 TOP 10 DES CONTRIBUABLES PAR MONTANT RECOUVRÉ:")
            print("-" * 50)
            
            total_top10 = 0
            for i, (contrib_id, montant) in enumerate(top_contribuables.items(), 1):
                # Formatage intelligent des montants
                if montant >= 1_000_000_000:
                    montant_fmt = f"{montant/1_000_000_000:.2f} Mds FCFA"
                elif montant >= 1_000_000:
                    montant_fmt = f"{montant/1_000_000:.2f} M FCFA"
                elif montant >= 1_000:
                    montant_fmt = f"{montant/1_000:.2f} K FCFA"
                else:
                    montant_fmt = f"{montant:.2f} FCFA"
                
                print(f"{i:2d}. {contrib_id:<15} | {montant_fmt:>15}")
                total_top10 += montant
            
            # Statistiques complémentaires
            total_general = df_work['MONTANT_RECOUVRE'].sum()
            pourcentage_top10 = (total_top10 / total_general) * 100 if total_general > 0 else 0
            
            print(f"\n📈 STATISTIQUES:")
            print(f"• Total Top 10: {total_top10:,.0f} FCFA")
            print(f"• Total général: {total_general:,.0f} FCFA")
            print(f"• Part du Top 10: {pourcentage_top10:.2f}%")
            
            # Informations détaillées sur le TOP 1
            if len(top_contribuables) > 0:
                top1_id = top_contribuables.index[0]
                top1_data = df_work[df_work['ID_CONTRIBUABLE'] == top1_id]
                
                print(f"\n🥇 DÉTAILS DU TOP 1 ({top1_id}):")
                if 'LIBELLE' in top1_data.columns:
                    libelles = top1_data['LIBELLE'].dropna().unique()[:3]
                    print(f"• Activités: {', '.join(libelles)}")
                if 'BUREAU' in top1_data.columns:
                    bureaux = top1_data['BUREAU'].dropna().unique()
                    print(f"• Bureau(x): {', '.join(bureaux)}")
                if 'DIRECTION' in top1_data.columns:
                    directions = top1_data['DIRECTION'].dropna().unique()
                    print(f"• Direction(s): {', '.join(directions)}")
                
                nb_transactions = len(top1_data)
                print(f"• Nombre de transactions: {nb_transactions:,}")
            
            return top_contribuables
            
        else:
            print("❌ Colonne MONTANT_RECOUVRE non trouvée!")
            
            # Auto-correction: Essayer avec d'autres colonnes
            print("🛠️  AUTO-CORRECTION: Recherche d'alternatives...")
            
            if 'MONTANT_DECLARE' in df_srmt.columns:
                print("✅ Utilisation de MONTANT_DECLARE comme alternative")
                df_work = df_srmt.copy()
                df_work['MONTANT_DECLARE'] = pd.to_numeric(df_work['MONTANT_DECLARE'], errors='coerce')
                df_work['MONTANT_DECLARE'] = df_work['MONTANT_DECLARE'].fillna(0)
                
                top_contribuables = (df_work.groupby('ID_CONTRIBUABLE')['MONTANT_DECLARE']
                                   .sum()
                                   .sort_values(ascending=False)
                                   .head(10))
                
                print("\n🏆 TOP 10 PAR MONTANT DÉCLARÉ (alternative):")
                for i, (contrib_id, montant) in enumerate(top_contribuables.items(), 1):
                    montant_fmt = f"{montant/1_000_000:.2f} M FCFA" if montant >= 1_000_000 else f"{montant/1_000:.1f} K FCFA"
                    print(f"{i:2d}. {contrib_id:<15} | {montant_fmt:>15}")
                    
                return top_contribuables
            
            else:
                raise Exception("Aucune colonne de montant appropriée trouvée")
    
    except Exception as e:
        print(f"❌ ERREUR FINALE: {str(e)}")
        print("🔍 Analyse des colonnes disponibles pour diagnostic:")
        for col in df_srmt.columns:
            sample_values = df_srmt[col].dropna().head(3).tolist()
            print(f"• {col}: {sample_values}")
        return None

# Exécuter l'analyse auto-corrective
resultat = analyser_top_contribuables_auto_corrective()

if resultat is not None:
    print(f"\n✅ Analyse terminée avec succès - {len(resultat)} contribuables identifiés")
else:
    print(f"\n❌ Échec de l'analyse - Vérifiez la structure des données")
```
